```python
import pandas as pd
pokemon=pd.read_csv("data/pokemon.csv",index_col="name")
pokemon.head()
```

```python
class Pokemon:
    def __init__(self,name,type,moves,evolutions,points_of_health='================'):
        #saves a variables as atributtes
        self.name= name
        self.type=type
        self.moves=moves
        self.attack=evolutions['attack']
        self.defense=evolutions['defense']
        self.points_of_health=points_of_health
        self.bar=20 # amount of points_of_health
        
        #pokemonName= Pokemon(pokemon_name,ivysaur_type, ivysaur_Abilitie, {'attack': ivysaur_SP_attack, 'defense':ivysaur_SP_defense})
```

```python


poke_dex=[]

for i in range(0,2):
    pokemon_name=input("Pokemon name: ") #    self.name= name
    salvage_Abilitie=pokemon.abilities[pokemon_name] # self.moves=moves
    salvage_type=pokemon.type1[pokemon_name] #  self.type=type
    salvage_SP_attack=pokemon.sp_attack[pokemon_name] # self.attack=evolutions['attack']
    salvage_SP_defense=pokemon.sp_defense[pokemon_name] # self.defense=evolutions['defense']
    salvage=Pokemon(pokemon_name,salvage_type, salvage_Abilitie, {'attack': salvage_SP_attack, 'defense':salvage_SP_defense})
    #poke_dex[i]=Pokemon(pokemon_name,ivysaur_type, ivysaur_Abilitie, {'attack': ivysaur_SP_attack, 'defense':ivysaur_SP_defense})
    poke_dex.append(salvage)
```
